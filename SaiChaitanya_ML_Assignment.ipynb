```python
import pandas as pd

df = pd.read_csv('/content/data.csv')

print("Basic Statistical Description:")
print(df.describe())

print("\nNull Values in DataFrame:")
print(df.isnull().sum())

df.fillna(df.mean(), inplace=True)
print("\nNull Values after Replacement:")
print(df.isnull().sum())

columns_to_aggregate = ['Pulse', 'Calories']
aggregated_data = df[columns_to_aggregate].agg(['min', 'max', 'count', 'mean'])
print("\nAggregated Data:")
print(aggregated_data)

filtered_df1 = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print("\nFiltered DataFrame (Calories between 500 and 1000):")
print(filtered_df1)

filtered_df2 = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print("\nFiltered DataFrame (calories values > 500 and pulse < 100):")
print(filtered_df2)

df_modified = df.drop(columns=['Maxpulse'])
print("\nModified DataFrame (without Maxpulse):")
print(df_modified)

df.drop(columns=['Maxpulse'], inplace=True)
print("\nMain DataFrame after deleting Maxpulse:")
print(df)

df['Calories'] = df['Calories'].astype(int)
print("\nData Types after Conversion:")
print(df.dtypes)
```
